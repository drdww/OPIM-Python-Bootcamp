## Unstructured Data
Text, images and time series are common in 'deep learning'. We won't cover these in too much detail this semester, but you should be aware that these data sets are only going to become more common for biz analytics and data scientists in the future!

### Text
Computers don't know how to read... but we can make text data look like structured (spreadsheet) data!


![one hot encoding](https://miro.medium.com/max/1600/1*nho6bONQ1qJoskvn89CmgA.png)

**Figure:** One-hot encoding. If the word is present in a sentence (document), it gets a 1 for each occurence, otherwise it gets a 0.

<center>

![tf-idf](https://plumbr.io/app/uploads/2016/06/tf-idf.png)

</center>

**Figure:** TF-IDF = 'term frequency - inverse document frequency' - rather than just one-hot encode, you also want to know if the word is unique to this sentence (document) vs. all other sentences you are.




<center>

![bag of words](https://www.researchgate.net/profile/Simon-Lin-9/publication/271910222/figure/fig1/AS:295071070081024@1447361833118/An-example-of-unigrams-bigrams-trigrams-and-4-grams-extracted-from-the-clinical-phrase.png)

</center>

**Figure:** we can try to get the order of the words by making bigrams and trigrams. Now you have phrases that you can run TF-IDF on! 

### Summary thoughts: text analytics
All of these text processing methods are great for machine learning (we'll learn other pre-processing like removing punctuation, making everything lowercase, removing common words (known as 'stop words'). In 'deep learning' we take it a step further and feed the ENTIRE SEQUENCE into the model. There's so much cool stuff to learn in analytics...

### Images
Images are known as 'unstructured data' - they aren't as easy to wrangle without advanced processing techniques. Although images are out of scope for this class (again, try 4+1 with BAPM!) - you should appreciate that images are data - they are numbers hiding in disguise!

![abe lincoln](https://miro.medium.com/max/1386/1*bV7S0zACdidh11ikjYpLpQ.png)

**Figure:** a pixel is realy just a number between 0 (dark, black) and 255 (light, white)! Computers really on these pixel values in order to find patterns in the data (automatically) for classification and regression problems. 

### Time Series
A lot of data businesses collect have a temporal component - this aspect can be exploited to make cool new products and tailored, customer analytics insights! As you can see below, we are predicting Google's stock price as a function of time. 

It used to be that you needed a lot of sophisticated time series and statistical knowledge in order to work with these data - with the advent of open-source tools, like 'FB Prophet' (yes, Facebook!), you can upload a time series and let the computer try the modeling for you.

```python
# psst... here is a sneak peak of the stuff you are going to do in my class...

# import modules
import pandas as pd
from pylab import *

# read and view some data (using pandas, making a dataframe)
df = pd.read_csv('sample_data/california_housing_train.csv')
df.head(n=5) # shows the first 5 rows in the dataset 
```

```python
# make a nice plot like the one we saw in the picture
scatter(x=df['median_income'], y=df['median_house_value'])
xlabel('Median Income')
ylabel('Median House Value')
title('House Value ~ Income') # always Y as a function of (~) X
show() # add this at the end so you don't have an annoying cryptic string...
```

```python
# make a nice plot like the one we saw in the picture
scatter(x=df['longitude'], y=df['latitude'])
xlabel('Longitude')
ylabel('Latitude')
title('Map of California') # always Y as a function of (~) X
show() # add this at the end so you don't have an annoying cryptic string...
```

It's not a *perfect* linear relationship, but there is definitely a trend there!!! When we explore a dataset for the first time, this is known as 'exploratory data analysis' or 'EDA'.
